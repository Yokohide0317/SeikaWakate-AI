```python
df = pd.read_csv("./data/GSE109607_norm_counts_TPM_GRCh38.p13_NCBI.tsv.gz", index_col=0, sep="\t")

sample_ids = {k: v for k, v in zip(df.columns[1:], """
GSM2947130	GSM2947131	GSM2947132	GSM2947133	GSM2947134	GSM2947135	GSM2947136	GSM2947137	GSM2947138	GSM2947139	GSM2947140	GSM2947141	GSM2947142	GSM2947143	GSM2947144	GSM2947145	GSM2947146	GSM2947147	GSM2947148	GSM2947149	GSM2947150	GSM2947151	GSM2947152	GSM2947153	GSM2947154	GSM2947155	GSM2947156	GSM2947157	GSM2947158	GSM2947159
""".strip().split("\t"))}

sample_names = """
GSM2947130	D-C1
GSM2947131	D-C2
GSM2947132	D-OPC1
GSM2947133	D-OPC2
GSM2947134	D-GSE1
GSM2947135	D-GSE2
GSM2947136	S6-C1
GSM2947137	S6-C2
GSM2947138	S6-OPC1
GSM2947139	S6-OPC2
GSM2947140	S6-GSE1
GSM2947141	S6-GSE2
GSM2947142	HCT-C1
GSM2947143	HCT-C2
GSM2947144	HCT-OPC1
GSM2947145	HCT-OPC2
GSM2947146	HCT-GSE1
GSM2947147	HCT-GSE2
GSM2947148	HT-C1
GSM2947149	HT-C2
GSM2947150	HT-OPC1
GSM2947151	HT-OPC2
GSM2947152	HT-GSE1
GSM2947153	HT-GSE2
GSM2947154	S4-C1
GSM2947155	S4-C2
GSM2947156	S4-OPC1
GSM2947157	S4-OPC2
GSM2947158	S4-GSE1
GSM2947159	S4-GSE2
""".strip().split("\n")
sample_names = [x.split("\t") for x in sample_names]
sample_names = {k: v for k, v in sample_names}

#df = df.rename(columns=sample_ids)
df.index = df.index.astype(str)
df
```

```python
entrezid = df.index.astype(str)
```

```python
%%R -i entrezid -o r_symbols -o r_emsembls

r_symbols <- select(
       x = org.Hs.eg.db,
       keys = entrezid,
       keytype = "ENTREZID",
       columns = c("SYMBOL"))

r_emsembls <- select(
       x = org.Hs.eg.db,
       keys = entrezid,
       keytype = "ENTREZID",
       columns = c("SYMBOL", "ENSEMBL"))
```

```python
symbols = r_symbols.copy().set_index("ENTREZID")
symbols.index = symbols.index.astype(str)

emsembls = r_emsembls.copy().set_index("ENTREZID")
emsembls.index = emsembls.index.astype(str)

symbols
```

```python
emsembls.drop(["SYMBOL"], axis=1).reset_index().drop_duplicates(subset="ENTREZID").set_index("ENTREZID")
```

```python
sample_df = df.copy()
sample_df = sample_df.merge(symbols, left_index=True, right_index=True, how="outer")

sample_df = sample_df.merge(
    emsembls.drop(["SYMBOL"], axis=1).reset_index().drop_duplicates(subset="ENTREZID").set_index("ENTREZID"),
    left_index=True, right_index=True, how="outer"
)

_index = []
for ind, row in sample_df.iterrows():
    if row["SYMBOL"] is not None:
        _index.append(row["SYMBOL"])
    elif row["SYMBOL"] is not None:
        _index.append(row["ENSEMBL"])
    else:
        _index.append(ind)

sample_df.index = _index
sample_df = sample_df.drop(["SYMBOL", "ENSEMBL"], axis=1)
sample_df.columns = sample_df.columns.map(lambda x: sample_names[x])

sample_df
```

```python
sample_df.to_csv("./data/bulkRNASeq_TPM-prep-SampleName.csv", index_label="Symbol")
```
