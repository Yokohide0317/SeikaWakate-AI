```python
metadata = pd.read_csv("./SraRunTable.txt", sep=",")
metadata = metadata[metadata["Assay Type"] == "AMPLICON"]
metadata = metadata[metadata["Sample Name"].str.contains("feces")]
metadata = metadata[metadata["Host_age"] == ">70"]
metadata = metadata[metadata["host_sex"] == "male"]

metadata = metadata.set_index("Run")
metadata = metadata[["Sample Name"]]
metadata
```

```python
metadata.sort_values("Sample Name")
```

```python
small_metadata = metadata[metadata["Sample Name"].str.contains("CTRL")].sample(10, random_state=17)
small_metadata = pd.concat([
    small_metadata, metadata[metadata["Sample Name"].str.contains("PDAC")].sample(10, random_state=17)
])
small_metadata
```

```python
small_metadata.drop(["Sample Name"], axis=1).to_csv("./SRR_Acc_List.txt", header=False, index=True)
```

```python
small_metadata.to_csv("./SraRunTable_sampled.csv", index=True)
```
